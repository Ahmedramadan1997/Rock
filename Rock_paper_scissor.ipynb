```python
import random

def Rock_paper_scissor():
    choices = ["Rock✊" , "paper✋" , "scissor✌️"]
    print("""winning rules of rock paper scissor game 
    Rock vs paper -> paper wins 
    paper vs scissor -> scissor wins 
    scissor vs Rock -> Rock wins """)
    print()
    print("""
    enter your choice
    1 -> Rock✊
    2-> paper✋
    3 -> scissor✌️""")
    print()
    while True:
        counter = 0
        user_score = 0
        pc_score = 0

        for i in range(3):

            pc_choices = random.choice(choices)
            user_choices = int(input(f"enter your choice :"))
            # user_choices :
            if user_choices == 1 :
                user_choices = "Rock✊"
            if  user_choices ==  2:
                user_choices = "paper✋"
            if  user_choices ==  3 :
                user_choices = "scissor✌️"
            print(f"the user_choices -> {user_choices} \nthe pc_choices -> {pc_choices}")

            if user_choices =="Rock✊" :
                if pc_choices == "scissor✌️" :
                    user_score += 1
                    print("congrat! you won this round\n")
                elif pc_choices == "paper✋" :
                    pc_score += 1
                    print("ooh you lost\n")
                else :
                    print("the round is draw\n")
            elif user_choices =="paper✋" :
                if pc_choices == "scissor✌️" :
                    pc_score += 1
                    print("ooh you lost\n")
                elif pc_choices == "Rock✊" :
                    user_score +=1
                    print("congrat! you won this round\n ")
                else :
                    print("the round is draw\n")         
            elif user_choices =="scissor✌️" :
                if pc_choices == "Rock✊" :
                    pc_score +=1
                    print("ooh you lost\n")
                elif pc_choices == "paper✋" :
                    user_score +=1
                    print("congrat! you won this round\n ")
                else :
                    print("the round is draw\n") 

        if  user_score > pc_score :
            print("congrats! you won the game")
        elif user_score < pc_score :
            print("you lost the game")
        else:
            print('the game tide')
        
        play_again = input("do you want to play again (y/n)").lower()
        
        if play_again == 'n':
            print('game over')
            break
        
```
